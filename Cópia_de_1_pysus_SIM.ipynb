# Download of the databases

```python
import pandas as pd
import numpy as np
from pysus.online_data import SIM
from pysus.online_data import SINAN, last_update
from pysus.online_data import CNES
```

## SIM

`CIRCOBITO`: Indicate the type of death:  
9: Ignored; 1: Accident; 2: Suicide;  3: Homicide; 4: Others; 

We downloaded only data from suicide and from the southern states = ParanÃ¡ (PR), Santa Catarina (SC), and Rio Grande do Sul (RS)

```python
df = pd.DataFrame()
south = ['PR', 'SC', 'RS']

for i in range(2009, 2021):
    for c in south:
        df1 = SIM.download(c, i)
        df1 = df1[df1['CIRCOBITO'] == '2']
        df = pd.concat([df, df1], ignore_index= True)

df
```

```python
df.to_csv('SUL_SIM.csv', index= False)
```

Since the data for "domestic violence" (VIOLBR21) for the year 2021 is not complete (stops in October), we collected data until 2020.

We downloaded only data from suicide and from the southern states = ParanÃ¡ (41), Santa Catarina (42), and Rio Grande do Sul (43) using the state code.

```python
df = pd.DataFrame()
for i in range(2009, 2022):
    df1 = SINAN.download('ViolÃªncia Domestica', i)
    PR = df1[df1['SG_UF_NOT'] == '41']
    SC = df1[df1['SG_UF_NOT'] == '42']
    RS = df1[df1['SG_UF_NOT'] == '43']
    south = pd.concat([PR, SC, RS], ignore_index=True)
    df2 = south[south['LES_AUTOP'] == '1']
    df = pd.concat([df, df2])
df
```

```python
df.to_csv('SUL_LOOP.csv', index= False)
```
